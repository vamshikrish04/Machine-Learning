```python
import pandas as pd
import matplotlib.pyplot as plt
#loading dataset
df = pd.read_csv('1.txt')
df.columns = ['population of a city','profit of a food truck']

def scatterplot(df, x_dim, y_dim):
   x = df[x_dim]
   y = df[y_dim]
   fig,ax = plt.subplots(figsize=(8, 5))
   ax.scatter(x, y)
#adds a title and axes labels
   ax.set_title('population vs profit of food')
   ax.set_xlabel('population ')
   ax.set_ylabel('profit of food (min)')
#removing top and right borders
   ax.spines['top'].set_visible(False)
   ax.spines['right'].set_visible(False)
   plt.show()
scatterplot(df,'population of a city','profit of a food truck')
```

# Major Gridlines
Something that I usually like to add to my plots are major gridlines. It helps with readability by reducing the amount of white background. You can play around with the its width linewidth and transparency alpha.


```python
import pandas as pd
import matplotlib.pyplot as plt
#loading dataset
df = pd.read_csv('1.txt')
df.columns = ['population of a city','profit of a food truck']

def scatterplot(df, x_dim, y_dim):
   x = df[x_dim]
   y = df[y_dim]
   fig,ax = plt.subplots(figsize=(8, 5))
   ax.scatter(x, y)
#adds a title and axes labels
   ax.set_title('population vs profit of food')
   ax.set_xlabel('population ')
   ax.set_ylabel('profit of food (min)')
#removing top and right borders
   ax.spines['top'].set_visible(False)
   ax.spines['right'].set_visible(False)
#adds major gridlines
   ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
   plt.show()
scatterplot(df,'population of a city','profit of a food truck')
```
